```python
"this is a script that outputs how many items one needs to prepare for a given number of guests"
```

Item list will include:
- fruit
- bread
- cheese
- drinks

```python

total = {
    'apples': 0,
    'bread': 0,
    'cheese': 0,
    'drinks': 0
}

print ('Are you coming to the picnic? (yes/no): ')
coming = input()

if coming.lower() == 'yes':
    print ('How many guests are you bringing? '+ "\n")
    guests = int(input())   
    
    for guestcount in range(1, guests + 1):
        print('For guest ' + str(guestcount) + ', write how many per item are needed:')
        print ('Apples: ')
        apples = int(input())
        print ('Bread: ')
        bread = int(input())
        print ('Cheese: ')
        cheese = int(input())
        print ('Drinks: '+ "\n")
        drinks = int(input())
    

        total['apples'] += apples
        total['bread'] += bread
        total['cheese'] += cheese
        total['drinks'] += drinks
        

    print('See you there!'+ "\n")
    print('Final Picnic Basket Total:')
    print('- Total Apples: ' + str(total['apples']))
    print('- Total Bread:  ' + str(total['bread']))
    print('- Total Cheese: ' + str(total['cheese']))
    print('- Total Drinks: ' + str(total['drinks']))

else:
    print("That's okay, maybe next time!")

```
