```python
import pandas as pd
from miRBench.dataset import get_dataset_path 

print("--- AGO2_CLASH_Hejret2023 Dataset ---")
path1 = get_dataset_path('AGO2_CLASH_Hejret2023', split="test")
df1 = pd.read_csv(path1, sep='\t', usecols=['gene', 'noncodingRNA', 'label'])
display(df1.head())

print("\n--- AGO2_eCLIP_Klimentova2022 Dataset ---")
path2 = get_dataset_path('AGO2_eCLIP_Klimentova2022', split="test")
df2 = pd.read_csv(path2, sep='\t', usecols=['gene', 'noncodingRNA', 'label'])
display(df2.head())
```

```python
unique_miRNA_df1 = pd.DataFrame(df1['noncodingRNA'].unique(), columns=['noncodingRNA'])
unique_miRNA_df2 = pd.DataFrame(df2['noncodingRNA'].unique(), columns=['noncodingRNA'])
unique_miRNA_df1.to_csv('unique_noncodingRNA_Hejret2023.csv', index=False)
unique_miRNA_df2.to_csv('unique_noncodingRNA_Klimentova2022.csv', index=False)
display(unique_miRNA1_df1.head())
display(unique_miRNA2_df2.head())

freq_hejret = {
    'Nucleotide': ['A', 'C', 'G', 'U'],  
    'Frequency': [                      
        unique_miRNA_df1['noncodingRNA'].str.count('A').sum(),
        unique_miRNA_df1['noncodingRNA'].str.count('C').sum(),
        unique_miRNA_df1['noncodingRNA'].str.count('G').sum(),
        unique_miRNA_df1['noncodingRNA'].str.count('U').sum()
    ]
}

freq_klimentova = {
    'Nucleotide': ['A', 'C', 'G', 'U'],  
    'Frequency': [                      
        unique_miRNA_df2['noncodingRNA'].str.count('A').sum(),
        unique_miRNA_df2['noncodingRNA'].str.count('C').sum(),
        unique_miRNA_df2['noncodingRNA'].str.count('G').sum(),
        unique_miRNA_df2['noncodingRNA'].str.count('U').sum()
    ]
}

df1_freq = pd.DataFrame(freq_hejret)
df2_freq = pd.DataFrame(freq_klimentova)
df1_freq.to_csv('Nucleotide freq per unique miRNA_Hejret2023.csv', index=False)
df2_freq.to_csv('Nucleotide freq per unique miRNA_Klimentova2022.csv', index=False)

print("\nNucleotide frequency per unique miRNA: AGO2_CLASH_Hejret2023")
display(df1_freq)

print("\nNucleotide frequency per unique miRNA: AGO2_eCLIP_Klimentova2022")
display(df2_freq)
```

K-mer count per unique miRNA

```python
import itertools

alphabet = ['A', 'C', 'G', 'U']
all_combinations = list(itertools.product(alphabet, repeat=3))
kmer_list = ["".join(item) for item in all_combinations]

combined_data = {
    '3-mer': kmer_list,
    
    'Hejret2023_Count': [
        sum(sum(1 for i in range(len(seq) - 2) if seq[i:i+3] == pattern) 
            for seq in unique_miRNA_df1['noncodingRNA'].dropna())
        for pattern in kmer_list
    ],
    
    'Klimentova2022_Count': [
        sum(sum(1 for i in range(len(seq) - 2) if seq[i:i+3] == pattern) 
            for seq in unique_miRNA_df2['noncodingRNA'].dropna())
        for pattern in kmer_list
    ]
}

df_kmer_summary = pd.DataFrame(combined_data)
df_kmer_summary.to_csv('Kmer_counts_summary.csv', index=False)

print("\n--- Global 3-mer Counts Summary Table ---")
display(df_kmer_summary.head(10)) # Shows the first 10 rows as a preview

```
