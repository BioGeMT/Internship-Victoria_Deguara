```python
from miRBench.dataset import get_dataset_df 

print("--- AGO2_CLASH_Hejret2023 Dataset ---")
df1 = get_dataset_df('AGO2_CLASH_Hejret2023', split="test")[['gene', 'noncodingRNA', 'label']]
display(df1.head())

print("\n--- AGO2_eCLIP_Klimentova2022 Dataset ---")
df2 = get_dataset_df('AGO2_eCLIP_Klimentova2022', split="test")[['gene', 'noncodingRNA', 'label']]
display(df2.head())

```

Dataset statistics
- no. of examples
- no. of +ve and -ve
- min, max, mean and med length of gene and noncodingRNA col.
- Alphabet of gene and noncodingRNA col.
- Check for duplicates

```python
datasets = {"AGO2_CLASH_Hejret2023": df1,
    "AGO2_eCLIP_Klimentova2022": df2}

print('number of examples in each datasets')
for name, df in datasets.items():
    print(f'    Number of examples in {name} dataset: {len(df)}')

print('\nPositive and negative distribution in each dataset:')
for name, df in datasets.items():
    for label in [1,0]:
        print (f'     {name}:Label {label}: {(df['label']==label).sum()}')

print('\nSequence lengths for each dataset:')
for name, df in datasets.items():
    for col in ['gene','noncodingRNA']:
        print (f'\n    Minimum length of {col} in {name} is: {df[col].str.len().min()}')
        print (f'    Maximum length of {col} in {name} is: {df[col].str.len().max()}')
        print (f'    Mean length of {col} in {name} is: {df[col].str.len().mean()}')
        print (f'    Median length of {col} in {name} is: {df[col].str.len().median()}')

print('\nThe alphabet in each dataset is:')
for name, df in datasets.items():
    for col in ['gene', 'noncodingRNA']:
        print(f'{col} in {name} contains: {set("".join(df[col].dropna()))}')

print ('\nThe number of duplicates in each dataset is:')
for name, df in datasets.items():
    ctdf = df['gene'].str.cat(df['noncodingRNA'])
    print(f'Number of duplicates in {name} dataset are: {ctdf.duplicated().sum()}')
    
```

K-mer count per unique miRNA

```python
import itertools

alphabet = ['A', 'C', 'G', 'U']
all_combinations = list(itertools.product(alphabet, repeat=3))
kmer_list = ["".join(item) for item in all_combinations]

combined_data = {
    '3-mer': kmer_list,
    
    'Hejret2023_Count': [
        sum(sum(1 for i in range(len(seq) - 2) if seq[i:i+3] == pattern) 
            for seq in unique_miRNA_df1['noncodingRNA'].dropna())
        for pattern in kmer_list
    ],
    
    'Klimentova2022_Count': [
        sum(sum(1 for i in range(len(seq) - 2) if seq[i:i+3] == pattern) 
            for seq in unique_miRNA_df2['noncodingRNA'].dropna())
        for pattern in kmer_list
    ]
}

df_kmer_summary = pd.DataFrame(combined_data)
df_kmer_summary.to_csv('Kmer_counts_summary.csv', index=False)

print("\n--- Global 3-mer Counts Summary Table ---")
display(df_kmer_summary.head(10)) # Shows the first 10 rows as a preview

```
